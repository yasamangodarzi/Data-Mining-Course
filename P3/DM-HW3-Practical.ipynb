<div dir="rtl">
<p dir=rtl style="direction: rtl;text-align: justify;line-height:200%;font-family:vazir;font-size:medium">
برای پردازش بهتر دیتاها ابتدا سطر های تکراری را پاک میکنیم. سپس برای ویژگی های عددی ستون های خالی و از دست رفته را با میانگین پر میکنیم.برای ویژگی هایCategorical هم سطرکه با مقدار ؟ پر شده اند رابا مقدار nan پر میکنیم و سپس با مقدار بیش ترین مقدار امده برای اون ویژگی پر میکنیم .(Mod)
</p>
<p dir=rtl style="direction: rtl;text-align: justify;line-height:200%;font-family:vazir;font-size:medium">
برای تبدیل مقادیرCategorical دو روش داریم که مزایای هرروش به صورت زیر است: 
</p>
<p dir=rtl style="direction: rtl;text-align: justify;line-height:200%;font-family:vazir;font-size:medium">
Label Encoding: برای ویژگی‌هایی که دارای ترتیب یا اولویت هستند معمولاً مناسب است.به راحتی می‌تواند به کاهش ابعاد ویژگی‌ها و حجم داده‌ها منجر شود
</p>
 <p dir=rtl style="direction: rtl;text-align: justify;line-height:200%;font-family:vazir;font-size:medium">
One-Hot Encoding: برای ویژگی‌های دسته‌ای بدون ترتیب و بدون اولویت، معمولاً بهترین راه‌حل است مانند رنگ‌ها یا کشورها 
</p>
     <p dir=rtl style="direction: rtl;text-align: justify;line-height:200%;font-family:vazir;font-size:medium">
در اینجا از روش اول استفاده میکنیم .و در نهایت نرمالسازی انجام میدهیم</p>
</div>

```python
# Data Preprocessing
from sklearn import preprocessing

# emove duplicated row 
train_dataset=train_dataset.drop_duplicates()

# fill data
for column in train_dataset.columns:
    train_dataset.loc[:, column] = train_dataset[column].replace(" ?", np.nan)
train_dataset = train_dataset.copy()
train_dataset['Age'].fillna(train_dataset['Age'].mean(), inplace=True)
train_dataset['Final_Weight'].fillna(train_dataset['Final_Weight'].mean(), inplace=True)
train_dataset['Education-Number'].fillna(train_dataset['Education-Number'].mean(), inplace=True)
train_dataset['Capital_Gain'].fillna(train_dataset['Capital_Gain'].mean(), inplace=True)
train_dataset['Capital_Loss'].fillna(train_dataset['Capital_Loss'].mean(), inplace=True)
train_dataset['Hours-Per-Week'].fillna(train_dataset['Hours-Per-Week'].mean(), inplace=True)
train_dataset['Work_Class'].fillna(train_dataset['Work_Class'].mode()[0], inplace=True)
train_dataset['Education'].fillna(train_dataset['Education'].mode()[0], inplace=True)
train_dataset['Marital_Status'].fillna(train_dataset['Marital_Status'].mode()[0], inplace=True)
train_dataset['Occupation'].fillna(train_dataset['Occupation'].mode()[0], inplace=True)
train_dataset['Relationship'].fillna(train_dataset['Relationship'].mode()[0], inplace=True)
train_dataset['Race'].fillna(train_dataset['Race'].mode()[0], inplace=True)
train_dataset['Sex'].fillna(train_dataset['Sex'].mode()[0], inplace=True)
train_dataset['Native_Country'].fillna(train_dataset['Native_Country'].mode()[0], inplace=True)

# convert Categorical data
label_encoder = preprocessing.LabelEncoder() 
train_dataset['Work_Class']= label_encoder.fit_transform(train_dataset['Work_Class']) 
train_dataset['Education']= label_encoder.fit_transform(train_dataset['Education']) 
train_dataset['Occupation']= label_encoder.fit_transform(train_dataset['Occupation'])
train_dataset['Marital_Status']= label_encoder.fit_transform(train_dataset['Marital_Status'])
train_dataset['Relationship']= label_encoder.fit_transform(train_dataset['Relationship'])
train_dataset['Race']= label_encoder.fit_transform(train_dataset['Race'])
train_dataset['Native_Country']= label_encoder.fit_transform(train_dataset['Native_Country'])
train_dataset['Sex']= label_encoder.fit_transform(train_dataset['Sex'])
# normalization
item_list = ['Age', 'Final_Weight', 'Education-Number', 'Capital_Gain', 'Capital_Loss', 
             'Hours-Per-Week' , 'Work_Class' ,'Education' , 'Marital_Status' ,'Relationship' , 'Occupation',
            'Race' ,'Sex', 'Native_Country']
for item  in item_list:
    mean = train_dataset[item].mean() 
    std = train_dataset[item].std()
    train_dataset[item] = (train_dataset[item] - mean) / std

train_dataset
```

<div dir="rtl">
<p dir=rtl style="direction: rtl;text-align: justify;line-height:200%;font-family:vazir;font-size:medium">
مراحل بالا را برای دیتاست تست هم انجام میدهیم . زیرا برای ارزیابی مدل از ان استفاده میکنیم.
</p>
</div>

```python
# Data Preprocessing
from sklearn import preprocessing

# emove duplicated row 
test_dataset=test_dataset.drop_duplicates()


# fill data
for column in test_dataset.columns:
    test_dataset.loc[:, column] = test_dataset[column].replace(" ?", np.nan)
test_dataset = test_dataset.copy()
test_dataset['Age'].fillna(test_dataset['Age'].mean(), inplace=True)
test_dataset['Final_Weight'].fillna(test_dataset['Final_Weight'].mean(), inplace=True)
test_dataset['Education-Number'].fillna(test_dataset['Education-Number'].mean(), inplace=True)
test_dataset['Capital_Gain'].fillna(test_dataset['Capital_Gain'].mean(), inplace=True)
test_dataset['Capital_Loss'].fillna(test_dataset['Capital_Loss'].mean(), inplace=True)
test_dataset['Hours-Per-Week'].fillna(test_dataset['Hours-Per-Week'].mean(), inplace=True)
test_dataset['Work_Class'].fillna(test_dataset['Work_Class'].mode()[0], inplace=True)
test_dataset['Education'].fillna(test_dataset['Education'].mode()[0], inplace=True)
test_dataset['Marital_Status'].fillna(test_dataset['Marital_Status'].mode()[0], inplace=True)
test_dataset['Occupation'].fillna(test_dataset['Occupation'].mode()[0], inplace=True)
test_dataset['Relationship'].fillna(test_dataset['Relationship'].mode()[0], inplace=True)
test_dataset['Race'].fillna(test_dataset['Race'].mode()[0], inplace=True)
test_dataset['Sex'].fillna(test_dataset['Sex'].mode()[0], inplace=True)
test_dataset['Native_Country'].fillna(test_dataset['Native_Country'].mode()[0], inplace=True)

# convert Categorical data
label_encoder = preprocessing.LabelEncoder() 
test_dataset['Work_Class']= label_encoder.fit_transform(test_dataset['Work_Class'])
test_dataset['Education']= label_encoder.fit_transform(test_dataset['Education'])
test_dataset['Occupation']= label_encoder.fit_transform(test_dataset['Occupation'])
test_dataset['Marital_Status']= label_encoder.fit_transform(test_dataset['Marital_Status'])
test_dataset['Relationship']= label_encoder.fit_transform(test_dataset['Relationship'])
test_dataset['Race']= label_encoder.fit_transform(test_dataset['Race'])
test_dataset['Native_Country']= label_encoder.fit_transform(test_dataset['Native_Country'])
test_dataset['Sex']= label_encoder.fit_transform(test_dataset['Sex'])
# normalization
item_list = ['Age', 'Final_Weight', 'Education-Number', 'Capital_Gain', 'Capital_Loss', 
             'Hours-Per-Week' , 'Work_Class' ,'Education' , 'Marital_Status' ,'Relationship' , 'Occupation',
            'Race' ,'Sex', 'Native_Country']
for item  in item_list:
    mean = test_dataset[item].mean()
    std = test_dataset[item].std()
    test_dataset[item] = (test_dataset[item] - mean) / std

test_dataset
```

```python
x_train = train_dataset[['Age', 'Final_Weight', 'Education-Number', 'Capital_Gain', 'Capital_Loss', 
             'Hours-Per-Week' , 'Work_Class' ,'Education' , 'Marital_Status' ,'Relationship' , 'Occupation',
            'Race' ,'Sex', 'Native_Country']]
y_train = train_dataset['Income']
x_test= test_dataset[['Age', 'Final_Weight', 'Education-Number', 'Capital_Gain', 'Capital_Loss', 
             'Hours-Per-Week' , 'Work_Class' ,'Education' , 'Marital_Status' ,'Relationship' , 'Occupation',
            'Race' ,'Sex', 'Native_Country']]
y_test = test_dataset['Income']
```

<div dir="rtl">
<p dir=rtl style="direction: rtl;text-align: justify;line-height:200%;font-family:vazir;font-size:medium">
برای استفاده از svm ابتدا مدل ان را از کتابخانه sklearn  اضافه کرده و سپس مدل را برروی دیتا ها اموزش فیت میکنیم و بعد به اموزش مدل میپردازیم. در نهایت هم معیار های خواسته شده را محاسبه میکنیم.</p>
</div>

```python
from sklearn.svm import SVC
from sklearn.metrics import classification_report,confusion_matrix
svm = SVC()
svm.fit(x_train, y_train)
predict_result = svm.predict(x_train)
report = classification_report(y_train, predict_result, output_dict=True, zero_division=1)
conf_matrix = confusion_matrix(y_train, predict_result)
print(conf_matrix)
print(f"Train SVM accuracy : {report['accuracy']}")
print(f"Train SVM recall : {report['weighted avg']['recall']}")
print(f"Train SVM support : {report['weighted avg']['support']}")
print(f"Train SVM f1-score : {report['weighted avg']['f1-score']}")
predict_result = svm.predict(x_test)
report = classification_report(y_test, predict_result, output_dict=True, zero_division=1)
conf_matrix = confusion_matrix(y_test, predict_result)
print(conf_matrix)
print(f"Test SVM accuracy : {report['accuracy']}")
print(f"Test SVM recall : {report['weighted avg']['recall']}")
print(f"Test SVM support : {report['weighted avg']['support']}")
print(f"Test SVM f1-score : {report['weighted avg']['f1-score']}")
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report

K_neighbors_classifier = KNeighborsClassifier(n_neighbors=7)
K_neighbors_classifier.fit(x_train, y_train)
predict_result = K_neighbors_classifier.predict(x_train)
report = classification_report(y_train, predict_result, output_dict=True, zero_division=1)
conf_matrix = confusion_matrix(y_train, predict_result)
print(conf_matrix)
print(f"Train K_neighbors accuracy : {report['accuracy']}")
print(f"Train K_neighbors recall : {report['weighted avg']['recall']}")
print(f"Train K_neighbors support : {report['weighted avg']['support']}")
print(f"Train K_neighbors f1-score : {report['weighted avg']['f1-score']}")
predict_result = K_neighbors_classifier.predict(x_test)
report = classification_report(y_test, predict_result, output_dict=True, zero_division=1)
conf_matrix = confusion_matrix(y_test, predict_result)
print(conf_matrix)
print(f"Test K_neighbors accuracy : {report['accuracy']}")
print(f"Test K_neighbors recall : {report['weighted avg']['recall']}")
print(f"Test K_neighbors support : {report['weighted avg']['support']}")
print(f"Test K_neighbors f1-score : {report['weighted avg']['f1-score']}")
```

```python
from sklearn.naive_bayes import GaussianNB
naive_bayes = GaussianNB()
naive_bayes.fit(x_train, y_train)
predict_result = naive_bayes.predict(x_train)
report = classification_report(y_train, predict_result, output_dict=True, zero_division=1)
conf_matrix = confusion_matrix(y_train, predict_result)
print(conf_matrix)
print(f"Train naive bayes accuracy : {report['accuracy']}")
print(f"Train naive bayes recall : {report['weighted avg']['recall']}")
print(f"Train naive bayes support : {report['weighted avg']['support']}")
print(f"Train naive bayes f1-score : {report['weighted avg']['f1-score']}")
predict_result = naive_bayes.predict(x_test)
report = classification_report(y_test, predict_result, output_dict=True, zero_division=1)
conf_matrix = confusion_matrix(y_test, predict_result)
print(conf_matrix)
print(f"Test naive bayes accuracy : {report['accuracy']}")
print(f"Test naive bayes recall : {report['weighted avg']['recall']}")
print(f"Test naive bayes support : {report['weighted avg']['support']}")
print(f"Test naive bayes f1-score : {report['weighted avg']['f1-score']}")
```

<div dir="rtl">
<p dir=rtl style="direction: rtl;text-align: justify;line-height:200%;font-family:vazir;font-size:medium">
برای ارزیابی مدل می توان از فرمول ها و پارامتر های مختلف استفاده کرد که ما در اینجا به ان ها میپردازیم ولی قبل از ان با ۴ اصطلاح باید اشنا شویم:
</p>
<p dir=rtl style="direction: rtl;text-align: justify;line-height:200%;font-family:vazir;font-size:medium">
TP (True Positive): تعداد نمونه هایی که به درستی تشخیص داده شده اند و واقعاً به کلاس مثبت تعلق دارند.
     <br>
TN (True Negative): تعداد نمونه هایی که به درستی تشخیص داده شده اند و واقعاً به کلاس منفی تعلق دارند.
     <br>
FP (False Positive): تعداد نمونه هایی که اشتباهاً به کلاس مثبت تشخیص داده شده اند، اما واقعاً به کلاس منفی تعلق دارند.
     <br>
FN (False Negative): تعداد نمونه هایی که اشتباهاً به کلاس منفی تشخیص داده شده اند، اما واقعاً به کلاس مثبت تعلق دارند.
</p>
 
<p dir=rtl style="direction: rtl;text-align: justify;line-height:200%;font-family:vazir;font-size:medium">
معیار اول ما Accuracy است و برابر است با :
<br>Accuracy=تعدادنمونه های درست/تعداد کل نمونه ها

</p>
<p dir=rtl style="direction: rtl;text-align: justify;line-height:200%;font-family:vazir;font-size:medium">
معیار دوم  ما Precision است و برابر است با :
    <br>Precision = TP/TP+FP

</p>
<p dir=rtl style="direction: rtl;text-align: justify;line-height:200%;font-family:vazir;font-size:medium">
معیار سوم  ما Recall است و برابر است با :
    <br>Recall = TP/TP+FN

</p>
<p dir=rtl style="direction: rtl;text-align: justify;line-height:200%;font-family:vazir;font-size:medium">
معیار چهارم  ما F1-Score  است و برابر است با :
    <br>F1-Score  = 2× (Precision×Recall/Precision+Recall)
</p>
    
<p dir=rtl style="direction: rtl;text-align: justify;line-height:200%;font-family:vazir;font-size:medium">
معیار اخر  ما Confusion Matrix 
    بصورت یک جدول نمایش داده می شود 
ماتریس به صورت زیر است:
     <br>
[TP FN
 <br>FP TN]
    <br>

</p>
<p dir=rtl style="direction: rtl;text-align: justify;line-height:200%;font-family:vazir;font-size:medium">
در نهایت مشاهده میکنیم که مدل های ما دچار OVERFITING  شده اند . به عبارت دیگر مدل های ما بیش از حد به داده های اموزش ما وابسته شده اندو نتوانسته اند پاسخگویی مناسبی نسبت به داده های تست باشند.به طور کلی می توان گفت جنگل تصادفی از همه مدل ها دقت بیش تری نسبت به داده های اموزش داشته است.
    برای کاهش بیش بردازش باید مدل را عوض کنیم یا تا حد امکان وابستگی را کاهش دهیم.
